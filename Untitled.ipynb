```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
import scipy.integrate as intg

def f_8(x):
    s1 = 3.7 * np.cos(1.5 * x) * np.exp(-4*x/ 3)
    s2 = 2.4 * np.sin(4.5 * x) * np.exp(2*x/3)
    return s1 + s2 + 4
# 
def p_8(x, a=1.8, b=2.3, alpha=0, beta=0.6):
    result = 1.0
    if alpha != 0:
        result *= (x - a)**(-alpha)
    if beta != 0:
        #   x = b!
        if abs(x - b) < 1e-12:  #  x   b
            return 0.0  #   
        result *= (b - x)**(-beta)
    return result

def F_8(x):
    return p_8(x, a = 1.8, b = 2.3, alpha = 0, beta = 0.6) * f_8(x)

ethalon_f_8 = intg.quad(f_8, a = 1.8, b = 2.3)[0]
ethalon_F_8 = intg.quad(F_8, a = 1.8, b = 2.3)[0]

#method
def left_rectangle(n, func = f_8, a = 1.8, b = 2.3):
    h = (b - a)/n
    total_sum = 0
    for i in range(0, n):
        x_i = a + i*h
        total_sum += func(x_i)
    return h*total_sum

def mid_rectangle(n, func = f_8, a = 1.8, b = 2.3):
    h = (b-a)/n
    total_sum = 0
    for i in range(0, n):
        x_i = a + (i + 0.5)*h
        total_sum += func(x_i)
    return h*total_sum

def trap_rectangle(n, func = f_8, a = 1.8, b = 2.3):
    h = (b-a)/n
    total_sum = 0
    total = (func(a)+func(b))/2
    for i in range(1, n):
        x_i = a + i*h
        total_sum += func(x_i)
    return h*(total_sum + total)

def sim_rectangle(n, func = f_8, a = 1.8, b = 2.3):
    h = (b-a)/n
    total_sum_1 = 0
    total_sum_2 = 0
    total = func(a) + func(b)
    for i in range(1, n):
        x_i = a + (i)*h
        if i % 2 == 1:
            total_sum_1 += func(x_i)
        else:
            total_sum_2 += func(x_i)
    return h/3 * (total + 4*total_sum_1 + 2*total_sum_2)

def Newton_cotes(n, func=f_8, a=1.8, b=2.3, alpha=0, beta=0.6):
    total = 0
    for i in range(n):  # i  0  n-1
        h = (b - a) / n
        z0 = a + i*h
        z1 = a + (i + 1/2)*h
        z2 = a + (i + 1)*h
        
        mu0 = ((b - z0)**0.4 - (b - z2)**0.4) / 0.4
        mu1 = b * ((b - z0)**0.4 - (b - z2)**0.4)/0.4 - ((b - z0)**1.4 - (b - z2)**1.4)/1.4
        mu2 = b*b * ((b - z0)**0.4 - (b - z2)**0.4)/0.4 - 2*b*((b - z0)**1.4 - (b - z2)**1.4)/1.4 + ((b - z0)**2.4 - (b - z2)**2.4)/2.4

        a1 = (mu2 - mu1*(z1 + z2) + mu0*z1*z2) / ((z1 - z0)*(z2 - z0))
        a2 = -(mu2 - mu1*(z0 + z2) + mu0*z0*z2) / ((z1 - z0)*(z2 - z1))
        a3 = (mu2 - mu1*(z0 + z1) + mu0*z0*z1) / ((z2 - z0)*(z2 - z1))

        total += a1*func(z0) + a2*func(z1) + a3*func(z2)
        
    return total
    
'''def Newton_cotes(func=F_8, a=1.8, b=2.3, beta = 0.6, m, i):
    h = (b - a) / m
    z0 = a +i*h
    z1 = a+ (i+1/2)*h
    z2 = a + (i+1)*h
    #u=[]
    #or i in range(3):
    #   mu.append(mui(i,z0,z2,b))

    mu0 = ((b - z0)**0.4 - (b - z2)**0.4) / 0.4
    mu1 = b * ((b - z0)**0.4 - (b - z2)**0.4)/0.4 - ((b - z0)** 1.4 - (b - z2)**1.4)/1.4
    mu2  = b*b * ((b - z0)**0.4 - (b - z2)**0.4)/0.4 - 2*b*((b - z0)** 1.4 - (b - z2)**1.4)/1.4 + ((b - z0)** 2.4 - (b - z2)**2.4)/2.4


    a1 = (mu2 - mu1*(z1 + z2) + mu0*z1*z2) / ((z1 - z0)*(z2 - z0))
    a2 = -(mu2 - mu1*(z0 + z2) + mu0*z0*z2) / ((z1 - z0)*(z2 - z1))
    a3 = (mu2 - mu1*(z0 + z1) + mu0*z0*z1) / ((z2 - z0)*(z2 - z1))

    return a1*func(z0) + a2*func(z1) + a3*func(z2)

def weighted_integral(f, a, b, beta, m):
    I = 0
    for i in range(0,m):
        I+= (Newt_Cotes(f,a,b,beta,m,i))

    return I'''

def Gauss(n, func = f_8, a1 = 1.8, b1 = 2.3, alpha = 0, beta = 0.6):
    h = (b1-a1)/n
    total = 0
    def Cardano(p):
        a, b, c, d = p[3], p[2], p[1], p[0]
        
        q = b**3/(27*a**3) - b*c/(6*a*a) + d/(2*a)
        p_val = (3*a*c - b*b)/(9*a*a)
    
        assert (q*q + p_val*p_val*p_val) < 0, "   "
        assert p_val < 0, " ,   "
        assert q*q + p_val**3 <= 0, " "
        r = np.sign(q) * np.sqrt(np.abs(p_val))
         
        phi = np.arccos(q/r**3)
        y1 = -2*r*np.cos(phi/3)
        y2 = 2*r*np.cos((np.pi/3) - (phi/3))
        y3 = 2*r*np.cos((np.pi/3) + (phi/3))
        
        x1 = y1 - b/(3*a)
        x2 = y2 - b/(3*a)
        x3 = y3 - b/(3*a)
                
        return [x1, x2, x3]
        
    for i in range(1, n+1):
        z0 = a1 + (i-1)*h
        z2 = a1 + i*h
        moments = []
        for k in range(6):
            integral = ((b1 - z0)**(k-beta+1) - (b1 - z2)**(k - beta + 1))/(k-beta+1)
            moments.append(integral)

        A = np.array([[moments[k + j] for j in range(3)] for k in range(3)])
        bbb = np.array([-moments[3 + k] for k in range(3)])
        a_coeffs = np.linalg.solve(A, bbb)

        p = [a_coeffs[0], a_coeffs[1], a_coeffs[2], np.float64(1)]
        
        nodes = Cardano(p)

        A_nodes = np.array([[nodes[j]**k for j in range(3)] for k in range(3)])
        b_moments = np.array([moments[k] for k in range(3)])
        A_coeffs = np.linalg.solve(A_nodes, b_moments)
        
        result = A_coeffs[0]*func(b1 - nodes[0]) + A_coeffs[1]*func(b1 - nodes[1]) + A_coeffs[2]*func(b1 - nodes[2])
        total += result    
    return total

n_val = [i for i in range(4, 30, 2)]

plt.figure(figsize=(10, 5))

method_1_1 = [left_rectangle, mid_rectangle, trap_rectangle, sim_rectangle]
for method in method_1_1:
    errors = []
    for n in n_val:
        approx = method(n, func = f_8)
        error = abs(approx - ethalon_f_8)
        errors.append(error)
    plt.plot(n_val, errors, label=method.__name__)


method_1_2 = [Newton_cotes, Gauss]
for method in method_1_2:
    errors = []
    for n in n_val:
        approx = method(n, func = f_8)
        error = abs(approx - ethalon_F_8)
        errors.append(error)
    plt.plot(n_val, errors, label=method.__name__)

plt.yscale('log')
plt.legend()
plt.show()
```
